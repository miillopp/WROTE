```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Задаем две функции
def y1(x): return np.log10(x)
def y2(x): return x**2 - 6

# 2. Диапазон x (помним про ОДЗ логарифма x > 0)
x = np.linspace(0.1, 4, 1000)

plt.figure(figsize=(8, 8))
plt.plot(x, y1(x), label=r'$y = \lg x$', color='blue')
plt.plot(x, y2(x), label=r'$y = x^2 - 6$', color='green')

# 3. Настройка осей как в тетрадке
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(-7, 5) # Ограничим по Y для наглядности
plt.legend()
plt.title('Графический метод: пересечение $\lg x$ и $x^2 - 6$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Задаем функции
def y1(x): return np.sin(x)
def y2(x): return 1/x

# 2. Создаем два диапазона (чтобы не было линии через разрыв в x=0)
x_left = np.linspace(-10, -0.1, 500)
x_right = np.linspace(0.1, 10, 500)

plt.figure(figsize=(10, 6))

# Рисуем синус
plt.plot(np.concatenate([x_left, x_right]), y1(np.concatenate([x_left, x_right])), 
         label=r'$y = \sin x$', color='purple')

# Рисуем гиперболу (двумя частями)
plt.plot(x_left, y2(x_left), color='red', label=r'$y = 1/x$')
plt.plot(x_right, y2(x_right), color='red')

# 3. Настройка осей
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.ylim(-2, 2) # Гипербола и синус лучше видны в этом масштабе
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.title('Графический метод: пересечение $\sin x$ и $1/x$')

plt.show()
```
