1 задание

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
import pandas as pd

def f(x):
    return log10(x) - cos(x)


a, b = 0.1, 3
step = 0.01

interval = None
x = a

while x < b:
    if f(x) * f(x + step) < 0:
        interval = (x, x + step)
        break
    x += step



def bisection(a, b, eps=1e-4):
    while (b - a) / 2 > eps:
        c = (a + b) / 2
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

root = bisection(interval[0], interval[1])


x_vals = np.linspace(0.1, 3, 1000)
y_vals = np.log10(x_vals) - np.cos(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label='lg(x) - cos(x)')
plt.axhline(0)


plt.scatter(root, 0, color='red', s=80)
plt.text(root, 0, f'{root:.4f}', color='red')

plt.grid()
plt.legend()
plt.title("Задание 1 ")
plt.show()

print(f"Корень: x ≈ {root:.4f}")
```

2 задание

```python
def f(x):
    return x**2 - 3.2*x - 1


x = np.linspace(-2, 5, 1000)
y = f(x)


def newton(x0, eps=1e-4):
    while True:
        df = 2*x0 - 3.2
        x1 = x0 - f(x0)/df
        if abs(x1 - x0) < eps:
            return x1
        x0 = x1

root = newton(3)


plt.figure(figsize=(8,5))
plt.plot(x, y, label='x^2 - 3.2x - 1')
plt.axhline(0)

plt.scatter(root, 0, color='red', s=80)
plt.text(root, 0, f'{root:.4f}', color='red')

plt.grid()
plt.legend()
plt.title("Задание 2")
plt.show()

print(f"Корень: x ≈ {root:.4f}")
```

```python
import math 
def f(x):
    return 4*x - 3*log(x) - 4

# график
x = np.linspace(0.1, 3, 1000)
y = 4*x - 3*np.log(x) - 4

# итерации
def phi(x):
    return (4 + 3*math.log(x)) / 4

def simple_iter(x0, eps=1e-4):
    while True:
        x1 = phi(x0)
        if abs(x1 - x0) < eps:
            return x1
        x0 = x1

root = simple_iter(1)

# график
plt.figure(figsize=(8,5))
plt.plot(x, y, label='4x - 3ln(x) - 4')
plt.axhline(0)

plt.scatter(root, 0, color='red', s=80)
plt.text(root, 0, f'{root:.4f}', color='red')

plt.grid()
plt.legend()
plt.title("Задание 3")
plt.show()

print(f"Корень: x ≈ {root:.4f}")
print(f"Проверка: 4*{root:.4f} - 3ln({root:.4f}) = {4*root - 3*np.log(root):.6f} (должно быть 4)")
```

```python
def df(x):
    return 2/((x+7)*log(10)) - 5*cos(x)

def combined(a, b, eps=1e-6):
    data = []
    n = 1
    x = a
    
    while True:
        x_chord = x - f(x)*(b-x)/(f(b)-f(x))
        x_newton = x - f(x)/df(x)
        
        x_new = (x_chord + x_newton)/2
        diff = abs(x_new - x)
        
        data.append([n, x, x_new, diff])
        
        if diff < eps:
            return x_new, data
        
        x = x_new
        n += 1

root4, steps4 = combined(0.35, 0.36)

df4 = pd.DataFrame(steps4, columns=["n","x","x_new","diff"]).round(6)
print(df4)

print(f"\nОтвет: x ≈ {root4:.6f}")
```
