```python
import matplotlib.pyplot as plt
import numpy as np

# Параметры системы
beta = 2
t = 2
p1, p2 = -1, 1

# 1. Вычисление количества чисел
n_pos = (beta - 1) * (beta**(t-1)) * (p2 - p1 + 1)
total_numbers = 2 * n_pos + 1

print(f"Количество положительных чисел: {int(n_pos)}")
print(f"Общее количество чисел в системе (с 0): {int(total_numbers)}")

# 2. Генерация всех возможных чисел
numbers = {0.0}
for sign in [-1, 1]:
    for p in range(p1, p2 + 1):
        # Мантисса нормализованного числа: d1.d2...dt
        # d1 от 1 до beta-1, остальные от 0 до beta-1
        for d1 in range(1, beta):
            for d2 in range(beta):
                mantissa = d1 * (beta**0) + d2 * (beta**-1)
                # Нормализованная форма: (d1/beta + d2/beta^2) * beta^p
                val = sign * (d1 / beta + d2 / (beta**2)) * (beta**(p + 1))
                numbers.add(val)

sorted_nums = sorted(list(numbers))
print(f"Список всех чисел: {sorted_nums}")

# 3. Визуализация на числовой оси
plt.figure(figsize=(12, 2))
plt.hlines(0, min(sorted_nums)-0.5, max(sorted_nums)+0.5, colors='black', alpha=0.5)
plt.plot(sorted_nums, np.zeros_like(sorted_nums), 'ro', label='Числа системы')
for x in sorted_nums:
    plt.text(x, 0.05, f'{x:g}', rotation=45, ha='right', fontsize=9)

plt.title(f"Система чисел (beta={beta}, t={t}, P=[{p1},{p2}])")
plt.yticks([])
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
beta_2 = 10
t_2 = 3
p1_2 = -9
p2_2 = 9

# Минимальное положительное нормализованное число
# Формат: 0.d1d2d3... * beta^p, где d1 != 0
# x_min = 0.100 * 10^-9 = 10^-10
x_min = (1 / beta_2) * (beta_2**p1_2)

# Максимальное число
# x_max = 0.999 * 10^9
x_max = (1 - beta_2**(-t_2)) * (beta_2**p2_2)

print(f"Граница машинного нуля (x_min): {x_min:.1e}")
print(f"Максимальное число системы (x_max): {x_max:.1f}")
```
