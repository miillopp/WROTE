```python
import numpy as np
import matplotlib.pyplot as plt


def phi(x):
    return 0.4 - 0.5 * x  # φ'(x) = -0.5, условие выполняется


a, b = -1.0, 2.0
xs = np.linspace(a, b, 400)
ys_phi = phi(xs)
ys_diag = xs

plt.figure(figsize=(6, 6))
plt.plot(xs, ys_phi, label='y = φ(x)')
plt.plot(xs, ys_diag, label='y = x', linestyle='--')


x = 1.5          # начальное приближение x0
n_iter = 8       # количество итераций

for k in range(n_iter):
    x_next = phi(x)

  
    plt.plot([x, x], [x, x_next], color='red')
    plt.plot([x, x_next], [x_next, x_next], color='red')

    x = x_next 


plt.xlim(a, b)
plt.ylim(a, b)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Сходящийся итерационный процесс x_{k+1} = φ(x)')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', 'box')

plt.show()
```

```python
import math

# 1) Функция φ(x), задающая итерации
def phi(x):
    # ВАЖНО: сюда можно подставить ту φ(x), которую дают в учебнике
    return 1.0 / math.sin(x)

# 2) Уравнение f(x) для контроля результата
def f(x):
    return x * math.sin(x) - 1.0

# 3) Параметры метода
x0 = 1.0          # начальное приближение (можно поменять при необходимости)
eps = 1e-5        # требуемая точность
max_iter = 1000   # защита от бесконечного цикла

x_prev = x0
for i in range(1, max_iter + 1):
    x_next = phi(x_prev)

    # проверка условия остановки по разности двух итераций
    if abs(x_next - x_prev) < eps:
        print(f"Итераций: {i}")
        break

    x_prev = x_next
else:
    print("Метод не сошёлся за max_iter шагов")
```
