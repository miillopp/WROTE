```python
import pandas as pd
import numpy as np

# 1. Функция
def f(x):
    return x * np.sin(x) - 1 

# 2. Исходные данные
a = 2.7
b = 2.8
eps = 1e-4

data = [] # Список для хранения строк таблицы

# 3. Цикл итераций
while (b - a) / 2 > eps:
    c = (a + b) / 2
    f_a = f(a)
    f_c = f(c)
    
    # Записываем текущие значения в список
    data.append({
        "Итерация": len(data) + 1,
        "a": round(a, 6),
        "b": round(b, 6),
        "c (середина)": round(c, 6),
        "f(c)": round(f_c, 6),
        "Длина отрезка": round(b - a, 6)
    })
    
    # Выбор новой границы
    if f_a * f_c < 0:
        b = c
    else:
        a = c

# 4. Вывод в виде красивой таблицы Pandas
df = pd.DataFrame(data)
display(df)

print(f"\nИтоговый корень: {round((a + b) / 2, 4)}")
```
