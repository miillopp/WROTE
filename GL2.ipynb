Задание 2

```python
def f(x):
    return x - 10 * np.sin(x)

a, b = 2, 3

print(f'\nПоиск корня на [{a}, {b}] с точностью {0.001}:')
print('   a       b       c     f(c)    |b-a|')

while b - a > 0.001:
    c = (a + b) / 2
    print(f'{a:.4f}  {b:.4f}  {c:.4f}  {f(c):.4f}  {b-a:.4f}')
    
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c

print(f'\nКорень: x = {c:.4f}')
print(f'Проверка: f({c:.4f}) = {f(c):.4f}')
```

задание 3

```python
# Начинаем с какого-то значения (например, 2.5)
x = 2.5 
step = 0
# Будем повторять, пока разница между старым и новым не станет крошечной
while True:
    x_new = 10 * np.sin(x) # Будет расходиться, поэтому используем арксинус:
    x_new = np.arcsin(x / 10)
    
    if abs(x_new - x) < 0.000001:
        break
    x = x_new
    step += 1

print("Корень:", x)
print("Количество итераций:", step)
```

Задание 4

```python
a = 2
b = 3
x = 0

# Обычный цикл на 100 повторений для точности
for i in range(100):
    # Формула хорд
    f_a = a - 10 * np.sin(a)
    f_b = b - 10 * np.sin(b)
    x = a - f_a * (b - a) / (f_b - f_a)
    
    # Проверяем, какой кусок отбросить
    if (a - 10 * np.sin(a)) * (x - 10 * np.sin(x)) < 0:
        b = x
    else:
        a = x
        
    # Если результат уже достаточно точный
    if abs(x - 10 * np.sin(x)) < 0.000001:
        break

print("Корень методом хорд:", x)
```
